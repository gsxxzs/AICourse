# Exersize 8.2 Uniform Distributions
Generate uniformed random points and plots its histogram 


## Experiment 1

```python
import random 
import matplotlib.pyplot as plt #8.1
import numpy as np 
from scipy import stats

M = 100 # number of points 

#exersize 2
def generate_random_points(M): #where M is the number of points (the book used M)                    
    X = [] 
    for _ in range(M):
        X.append(random.uniform(0,1))
    
    y = np.empty(M)
    y.fill(0.5)
    point_size = 2 
    plt.scatter(X, y, color = "red", s = point_size)
    plt.axis([-0.25, 1.25, 0, 1]) #x and y range for plot 
    plt.title('%d Uniformly Randomly Generated Points' %M) 
    plt.show()
    
    plt.title("Histogram of Uniform Distribution")
    plt.hist(X,  bins = 30, edgecolor='black')
    plt.title("Histogram of Uniform Distribution")
    plt.show()
    return X
    
X = generate_random_points(M)
```

## Experiment 2
Transforms the uniformed distributed random points to a normal distribution centered a 0 with a standard deviation of 1

```python
result=stats.norm.ppf(X,loc=0,scale=1) #Uses inverse CDF method to make a normal distribution 反变换采样法

y = np.empty(M)
y.fill(0.5)
plt.scatter(result, y, color = "red", s = 2)
plt.axis([-4, 4, 0, 1])
plt.title('%d Uniformly Randomly Generated Points Transformed to Normal' %M) 
plt.show()
    
plt.hist(result,  bins = 10, edgecolor='black')
plt.title("Histogram of Uniform Distribution Transformed to Normal Distribution")

plt.show()
```
